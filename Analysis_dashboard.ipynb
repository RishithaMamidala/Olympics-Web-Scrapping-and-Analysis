```python
# total number of nations who participated in each olympics game?
pd.read_sql("""
SELECT 
    h.game,
    c.host_city,
    COUNT(DISTINCT h.noc) AS nb_nations
FROM olympics.athlete_data h
JOIN olympics.host_cities c ON h.game = c.game
JOIN olympics.athletes_roles r ON h.id = r.id
WHERE r.roles NOT LIKE '%(non-medal events)%' and r.roles NOT LIKE 'Non-starter'
GROUP BY h.game, c.host_city
order by h.game
""", conn)
```

```python
# Which year saw the highest and lowest number of countries participating in olympics?

pd.read_sql("""
with no_nations_cte as(
SELECT 
    h.game,
    c.host_city,
    COUNT(DISTINCT h.noc) AS nb_nations
FROM olympics.athlete_data h
JOIN olympics.host_cities c ON h.game = c.game
JOIN olympics.athletes_roles r ON h.id = r.id
WHERE r.roles NOT LIKE '%(non-medal events)%' and r.roles NOT LIKE 'Non-starter'
GROUP BY h.game, c.host_city)
            
SELECT DISTINCT
    CONCAT(FIRST_VALUE(game) OVER (ORDER BY nb_nations ASC), ' - ', 
        FIRST_VALUE(nb_nations) OVER (ORDER BY nb_nations ASC), ' nations') AS lowest_nb_nations,
    CONCAT(FIRST_VALUE(game) OVER (ORDER BY nb_nations DESC), ' - ', 
        FIRST_VALUE(nb_nations) OVER (ORDER BY nb_nations DESC), ' nations') AS highest_nb_nations
FROM no_nations_cte

""", conn)
```

```python
pd.read_sql("""
with no_games_nation_cte as
(
    select athlete.team, count(distinct athlete.game) as no_of_games_attended
    from olympics.athlete_data as athlete
    Inner Join olympics.host_cities as host_city
    on athlete.game = host_city.game 
    group by athlete.team
)

select team, no_of_games_attended from no_games_nation_cte
where no_of_games_attended = (
    select count(distinct athlete.game)
    FROM olympics.athlete_data as athlete
    Inner Join olympics.host_cities as host_city
    on athlete.game = host_city.game    
)
            
""", conn)
```

```python
# Identify the sport which was played in all summer olympics.
pd.read_sql("""
SELECT COUNT(DISTINCT h.game) AS "Total Number of Summer Games"
FROM olympics.athlete_data h
JOIN olympics.host_cities c ON h.game = c.game
WHERE c.season = 'Summer'
""", conn) 
```

```python
pd.read_sql("""
WITH count_summer_games_cte AS
(
    SELECT h.sport, COUNT(DISTINCT h.game) AS total_summer_games_attended
    FROM olympics.athlete_data h
    JOIN olympics.host_cities c ON h.game = c.game
    WHERE c.season = 'Summer'
    GROUP BY h.sport
)
SELECT sport, total_summer_games_attended
FROM count_summer_games_cte
WHERE total_summer_games_attended =
(
    SELECT COUNT(DISTINCT h.game) AS "Total Number of Summer Games"
    FROM olympics.athlete_data h
    JOIN olympics.host_cities c ON h.game = c.game
    WHERE c.season = 'Summer'
)
""", conn) 
```

```python
pd.read_sql("""
WITH count_games_cte AS (
    SELECT h.sport, COUNT(DISTINCT h.game) AS total_games_attended
    FROM olympics.athlete_data h
    JOIN olympics.host_cities c ON h.game = c.game
    GROUP BY h.sport
    HAVING COUNT(DISTINCT h.game) = 1
)
SELECT distinct cte.sport, h.game, c.host_city, cte.total_games_attended
FROM olympics.athlete_data h
JOIN olympics.host_cities c ON h.game = c.game
JOIN count_games_cte cte ON h.sport = cte.sport
order by h.game
""", conn) 
```

```python
#Fetch the total number of sports played in each olympic games.
pd.read_sql("""
SELECT 
    h.game, 
    c.host_city, 
    COUNT(DISTINCT h.sport) AS total_number_sports
FROM olympics.athlete_data h
JOIN olympics.host_cities c ON h.game = c.game
JOIN olympics.athletes_roles r ON h.id = r.id
WHERE r.roles NOT LIKE '%(non-medal events)%' and r.roles NOT LIKE 'Non-starter'
GROUP BY h.game, c.host_city
""", conn)
```

```python
#Fetch details of the oldest athletes to win a gold medal.
pd.read_sql("""
SELECT TOP 5
    h.id, 
    h.name, 
    h.birth_year,
    h.death_year,
    c.year,
    c.year - h.birth_year AS age,
    n.country,
    h.sport,
    COUNT(h.medal) AS gold_medals
    
FROM olympics.athlete_data h
JOIN olympics.countries_noc n ON h.noc = n.noc
JOIN olympics.host_cities c on h.game = c.game
WHERE 
    h.medal = 'Gold' 
    AND h.birth_year IS NOT NULL
    AND h.event LIKE '%Olympic%'
GROUP BY h.id, h.name, h.birth_year, h.death_year, n.country, h.sport,h.birth_year - c.year,c.year
ORDER BY age DESC

""", conn)
```

```python
#Fetch details of the youngest athletes to win a gold medal.
pd.read_sql("""
SELECT TOP 15
    h.id, 
    h.name, 
    h.birth_year,
    h.death_year,
    c.year,
    c.year - h.birth_year AS age,
    n.country,
    h.sport,
    COUNT(h.medal) AS gold_medals
    
FROM olympics.athlete_data h
JOIN olympics.countries_noc n ON h.noc = n.noc
JOIN olympics.host_cities c on h.game = c.game
WHERE 
    h.medal = 'Gold' 
    AND h.birth_year IS NOT NULL
    
GROUP BY h.id, h.name, h.birth_year, h.death_year, n.country, h.sport,h.birth_year - c.year,c.year
ORDER BY age 

""", conn)
```

```python
# Find the Ratio of male and female athletes participated in all olympic games.
pd.read_sql("""
SELECT 
    CONVERT(VARCHAR, SUM(CASE WHEN gender = 'Male' THEN 1 ELSE 0 END)) AS nb_males,
    CONVERT(VARCHAR, SUM(CASE WHEN gender = 'Female' THEN 1 ELSE 0 END)) AS nb_females,
    CONVERT(VARCHAR, COUNT(id)) AS total_athletes,
    FORMAT(SUM(CASE WHEN gender = 'Male' THEN 1 ELSE 0 END) * 100.0 / COUNT(id), '0.#') + ' %' AS males_percent,
FORMAT(SUM(CASE WHEN gender = 'Female' THEN 1 ELSE 0 END) * 100.0 / COUNT(id), '0.#') + ' %' AS females_percent,
FORMAT(SUM(CASE WHEN gender = 'Male' THEN 1 ELSE 0 END) * 1.0 / SUM(CASE WHEN gender = 'Female' THEN 1 ELSE 0 END), '0') + ' : 1' AS "Ratio M:F"  
FROM 
(
    -- Retrieve unique IDs            

SELECT DISTINCT h.id, h.gender, h.game
    FROM olympics.athlete_data h
    JOIN olympics.athletes_roles r ON h.id = r.id
    JOIN olympics.host_cities c on h.game = c.game
    WHERE 
        r.roles NOT LIKE '%(non-medal events)%' 
        AND r.roles NOT LIKE '%Referee%'
        AND r.roles NOT LIKE '%Administrator%'
        AND r.roles LIKE '%Competed in Olympic Games%' 
) unique_id;
""", conn)
```

```python
# In which year were female athletes permitted to participate in the Olympic Games? Provide a breakdown of the male and female participants for each Olympic Games.
pd.read_sql("""   
WITH unique_id_cte AS
(
    SELECT 
        DISTINCT id, 
        name, 
        gender, 
        game
    FROM 
    (
        SELECT h.id, h.name, h.gender, h.game
        FROM olympics.athlete_data h
        JOIN olympics.athletes_roles r ON h.id = r.id
        JOIN olympics.host_cities c on h.game = c.game 
        WHERE 
            r.roles NOT LIKE '%(non-medal events)%' 
            AND r.roles NOT LIKE '%Referee%'
            AND r.roles NOT LIKE '%Administrator%'
            AND r.roles LIKE '%Competed in Olympic Games%' 
    ) e
)
SELECT 
    game,
    SUM(CASE WHEN gender = 'Male' THEN 1 ELSE 0 END) AS nb_males,
    SUM(CASE WHEN gender = 'Female' THEN 1 ELSE 0 END) AS nb_females,
    COUNT(id) AS total_athletes
FROM unique_id_cte
GROUP BY game
ORDER BY game
""", conn)
```

```python
#Fetch the top 5 athletes who have won the most gold medals.
pd.read_sql("""
SELECT TOP 5    
    h.id, 
    h.name, 
    n.country, 
    COUNT(h.medal) AS gold_medals
FROM olympics.athlete_data h
JOIN olympics.countries_noc n ON h.noc = n.noc
JOIN olympics.host_cities c on h.game = c.game 
WHERE h.event LIKE '%Olympic%' AND h.medal = 'Gold'
GROUP BY h.id, h.name, n.country, h.noc
ORDER BY gold_medals DESC, n.country DESC
""", conn)
```

```python
#Fetch the top 5 athletes who have won the most medals (gold/silver/bronze).
pd.read_sql("""
SELECT TOP 5
    h.id, 
    h.name, 
    n.country, 
    SUM(CASE WHEN h.medal = 'Gold' THEN 1 ELSE 0 END) AS gold,
    SUM(CASE WHEN h.medal = 'Silver' THEN 1 ELSE 0 END) AS silver,
    SUM(CASE WHEN h.medal = 'Bronze' THEN 1 ELSE 0 END) AS bronze,
    SUM(CASE WHEN h.medal = 'Gold' THEN 1 WHEN h.medal = 'Silver' THEN 1 WHEN h.medal = 'Bronze' THEN 1 ELSE 0 END) AS total_medals
FROM olympics.athlete_data h
JOIN olympics.countries_noc n ON h.noc = n.noc
JOIN olympics.host_cities c on h.game = c.game 
WHERE h.event LIKE '%Olympic%'
GROUP BY h.id, h.name, n.country
ORDER BY total_medals DESC, gold DESC
""", conn)
```

```python
#Fetch the top 5 most successful countries in olympics. Success is defined by no of medals won.
pd.read_sql("""
SELECT TOP 5
    n.country,
    SUM(CASE WHEN h.medal = 'Gold' THEN 1 WHEN h.medal = 'Silver' THEN 1 WHEN h.medal = 'Bronze' THEN 1 ELSE 0 END) AS total_medals
FROM olympics.athlete_data h
JOIN olympics.countries_noc n ON h.noc = n.noc
JOIN olympics.host_cities c on h.game = c.game 
WHERE h.event LIKE '%Olympic%'
GROUP BY n.country
ORDER BY total_medals DESC
""", conn)
```

```python
pd.read_sql("""
WITH count_medals_cte AS
(
    -- Partition the dataset by country, specific games, events and medals
    
    SELECT 
        n.country, 
        h.game,
        h.event, 
        h.medal,
        ROW_NUMBER() OVER (PARTITION BY n.noc, h.game, h.event, h.medal order by h.medal) AS rk
    FROM olympics.athlete_data h
    JOIN olympics.countries_noc n ON h.noc = n.noc
    JOIN olympics.host_cities c on h.game = c.game 
    WHERE h.event LIKE '%Olympic%' and
    h.medal NOT LIKE 'NULL'
)
SELECT TOP 5 country, COUNT(medal) AS total_medals
FROM count_medals_cte
WHERE rk = 1
GROUP BY country
ORDER BY total_medals DESC

""", conn)
```

```python
#List down total gold, silver and broze medals won by each country corresponding to each olympic games
pd.read_sql("""
WITH count_medals_cte AS
(
    -- Partition the dataset by country, specific games, events and medals
    
    SELECT
        h.game,
        n.country,
        h.event, 
        h.medal,
        ROW_NUMBER() OVER (PARTITION BY h.noc, n.country, h.game, h.event, h.medal order by h.medal) AS rk
    FROM olympics.athlete_data h
    JOIN olympics.countries_noc n ON h.noc = n.noc
    JOIN olympics.host_cities c on h.game = c.game 
    WHERE h.event LIKE '%Olympic%'  and
    h.medal NOT LIKE 'NULL'
 
)
SELECT DISTINCT TOP 10
    game,
    country,
    SUM(CASE WHEN medal = 'Gold' THEN 1 ELSE 0 END) AS gold,
    SUM(CASE WHEN medal = 'Silver' THEN 1 ELSE 0 END) AS silver,
    SUM(CASE WHEN medal = 'Bronze' THEN 1 ELSE 0 END) AS bronze,
    COUNT(medal) AS total_medals
FROM count_medals_cte 
WHERE rk = 1
GROUP BY game, country
ORDER BY game 
""", conn)
```

```python
#Identify which country won the most gold, most silver, most bronze medals and the most medals in each olympic games.
pd.read_sql("""
WITH count_medals_cte AS
(
    -- Partition the dataset by country, specific games, events and medals
    
    SELECT
        h.game,
        n.country,
        h.event, 
        h.medal,
        ROW_NUMBER() OVER (PARTITION BY h.noc, n.country, h.game, h.event, h.medal order by h.medal) AS rk
    FROM olympics.athlete_data h
    JOIN olympics.countries_noc n ON h.noc = n.noc
    JOIN olympics.host_cities c on h.game = c.game 
    WHERE h.event LIKE '%Olympic%'  and
    h.medal NOT LIKE 'NULL'
 
),
count_type_medals_cte AS
(
    SELECT
        game,
        country,
        SUM(CASE WHEN medal = 'Gold' THEN 1 ELSE 0 END) AS gold,
        SUM(CASE WHEN medal = 'Silver' THEN 1 ELSE 0 END) AS silver,
        SUM(CASE WHEN medal = 'Bronze' THEN 1 ELSE 0 END) AS bronze,
        COUNT(medal) AS total_medals
    FROM count_medals_cte
    WHERE rk = 1
    GROUP BY game, country

)
SELECT DISTINCT
    game,
    CONCAT(
        FIRST_VALUE(country) OVER (PARTITION BY game ORDER BY gold DESC), ' - ',
        FIRST_VALUE(gold) OVER (PARTITION BY game ORDER BY gold DESC)
    ) AS most_gold,
    
    CONCAT(
        FIRST_VALUE(country) OVER (PARTITION BY game ORDER BY silver DESC), ' - ',
        FIRST_VALUE(silver) OVER (PARTITION BY game ORDER BY silver DESC)
    ) AS most_silver,
        
    CONCAT(
        FIRST_VALUE(country) OVER (PARTITION BY game ORDER BY bronze DESC), ' - ',
        FIRST_VALUE(bronze) OVER (PARTITION BY game ORDER BY bronze DESC)
    ) AS most_bronze,
    
    CONCAT(
        FIRST_VALUE(country) OVER (PARTITION BY game ORDER BY total_medals DESC), ' - ',
        FIRST_VALUE(total_medals) OVER (PARTITION BY game ORDER BY total_medals DESC)
    ) AS most_medals
    
FROM count_type_medals_cte 
ORDER BY game
""", conn)
```

```python
#Which countries have never won gold medal but have won silver/bronze medals?
pd.read_sql("""
WITH count_medals_cte AS
(
    -- Partition the dataset by country, specific games, events and medals
    
    SELECT
        n.country,
        h.game,
        h.event, 
        h.medal,
        ROW_NUMBER() OVER (PARTITION BY h.noc, n.country, h.game, h.event, h.medal order by h.medal) AS rk
    FROM olympics.athlete_data h
    JOIN olympics.countries_noc n ON h.noc = n.noc
    JOIN olympics.host_cities c on h.game = c.game 
    WHERE h.event LIKE '%Olympic%'  and
    h.medal NOT LIKE 'NULL'
)
SELECT 
    country,
    SUM(CASE WHEN medal = 'Gold' THEN 1 ELSE 0 END) AS gold,
    SUM(CASE WHEN medal = 'Silver' THEN 1 ELSE 0 END) AS silver,
    SUM(CASE WHEN medal = 'Bronze' THEN 1 ELSE 0 END) AS bronze,
    COUNT(medal) AS total_medals
FROM count_medals_cte
WHERE rk = 1
GROUP BY country
HAVING 
    SUM(CASE WHEN medal = 'Gold' THEN 1 ELSE 0 END) = 0 AND 
    (SUM(CASE WHEN medal = 'Silver' THEN 1 ELSE 0 END) > 0 OR
    SUM(CASE WHEN medal = 'Bronze' THEN 1 ELSE 0 END) >0)
ORDER BY country 
""", conn)
```
